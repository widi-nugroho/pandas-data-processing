## Understanding Joins

### Create a DataFrame

```python
product=pd.DataFrame({
    'Product_ID':[101,102,103,104,105,106,107],
    'Product_name':['Watch','Bag','Shoes','Smartphone','Books','Oil','Laptop'],
    'Category':['Fashion','Fashion','Fashion','Electronics','Study','Grocery','Electronics'],
    'Price':[299.0,1350.50,2999.0,14999.0,145.0,110.0,79999.0],
    'Seller_City':['Delhi','Mumbai','Chennai','Kolkata','Delhi','Chennai','Bengalore']
})
product
```

```python
customer=pd.DataFrame({
    'id':[1,2,3,4,5,6,7,8,9],
    'name':['Olivia','Aditya','Cory','Isabell','Dominic','Tyler','Samuel','Daniel','Jeremy'],
    'age':[20,25,15,10,30,65,35,18,23],
    'Product_ID':[101,0,106,0,103,104,0,0,107],
    'Purchased_Product':['Watch','NA','Oil','NA','Shoes','Smartphone','NA','NA','Laptop'],
    'City':['Mumbai','Delhi','Bangalore','Chennai','Chennai','Delhi','Kolkata','Delhi','Mumbai']
})
customer
```

### Merge 2 DataFrame where the "Product_ID" exist on both DataFrame

```python
product_customer=pd.merge(product,customer,on='Product_ID')
product_customer
```

### Merge 2 DataFrame 

```python
product_customer=pd.merge(product,customer,on='Product_ID',how='outer')
product_customer
```

### Merge 2 DataFrame with all the product_id taken from the Product DataFrame

```python
product_customer=pd.merge(product,customer,on='Product_ID',how='left')
product_customer
```

### Merge 2 DataFrame with all the product_id taken from the Customer DataFrame

```python
product_customer=pd.merge(product,customer,on='Product_ID',how='right')
product_customer
```

## Handling Redundancy/Duplicates in Joins

```python
product_dup=pd.DataFrame({
'Product_ID':[101,102,103,104,105,106,107,103,107],
'Product_name':['Watch','Bag','Shoes','Smartphone','Books','Oil','Laptop','Shoes','Laptop'],
'Category':['Fashion','Fashion','Fashion','Electronics','Study','Grocery','Electronics','Fashion','Electronics'],
'Price':[299.0,1350.50,2999.0,14999.0,145.0,110.0,79999.0,2999.0,79999.0],
'Seller_City':['Delhi','Mumbai','Chennai','Kolkata','Delhi','Chennai','Bengalore','Chennai','Bengalore']
})
product_dup              
```

### Merge 2 DataFrame with duplicates

```python
product_consumer=pd.merge(product_dup,customer,how='inner',on='Product_ID')
product_consumer
```

### Merge 2 DataFrame without duplicates

```python
product_consumer=pd.merge(product_dup.drop_duplicates(),customer,how='inner',on='Product_ID')
product_consumer
```

there is a validate argument in the merge() function, which we can set to â€˜one_to_oneâ€™, â€˜one_to_manyâ€™, â€˜many_to_oneâ€™, and â€˜many_to_manyâ€™.

### Merge 2 DataFrame with validate argument

```python
product_consumer=pd.merge(product_dup,customer,how='inner',on='Product_ID',validate='many_to_many')
product_consumer
```
