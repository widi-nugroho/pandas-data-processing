# Grouping and Sorting

### import libraries and load data

```python
import pandas as pd
import numpy as np
reviews = pd.read_csv("winemag-data-130k-v2.csv",index_col=0)
reviews.head(2)
```

## Groupwise analysis

### Group by a columns counts

```python
reviews.groupby('points').points.count()
```

### Group "points" by minimal "price"

```python
reviews.groupby('points').price.min()
```

### Group "points" by average "price"

```python
reviews.groupby('points').price.mean()
```

### Group "winery" and assign the first title of each winery to the group

```python
grouped_reviews=reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
grouped_reviews
```

### Single index grouping has a Series type

```python
type(grouped_reviews)
```

### Group by multiple index

```python
#idxmax() :Return index of first occurrence of maximum over requested axis.
grouped_reviews=reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
grouped_reviews.head(5)
```

### Multiple index grouping has a DataFrame type

```python
type(grouped_reviews)
```

Another groupby() method worth mentioning is agg(), which lets you run a bunch of different functions on your DataFrame simultaneously. For example, we can generate a simple statistical summary of the dataset as follows:

### Group with a bunch of diferent functions

```python
#Group country by (total price of each country, minimal price of each country, maximal price of each country)
grouped_reviews=reviews.groupby(['country']).price.agg([len, min, max])
grouped_reviews
```

### Check the grouped_reviews type

```python
type(grouped_reviews)
```

## Multi-indexes grouping

```python
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

### The index type of multi-index grouping is multi index

```python
mi = countries_reviewed.index
mi[0]
```

```python
type(mi)
```

### Convert to a regular index

```python
countries_reviewed.reset_index()
```

## Sorting

### Sort DataFrame by a numerical column (ascending)

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

### Sort DataFrame by a numerical column (Descending)

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

### Sort by index

```python
countries_reviewed.sort_index()
```

### Sort by multiple values

```python
countries_reviewed.sort_values(by=['country', 'len'])
```
